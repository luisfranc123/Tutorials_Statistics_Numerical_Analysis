When plotting in three dimensions, sometimes it is desirable to use a surface plot rather than a line plot. In 3D surface plotting, we wish to make a graph of some relationship $f(x, y)$. In surface plotting all $(x, y)$ pairs must be given. This is not straightforward to do using vectors; therefore, in surface plotting, the first data structure we must create is called a mesh. Given lists/arrays of $x$ and $y$ values, a mesh is a list of possible combinations of $x$ and $y$. A third array, `Z`, can then be created such that `Z[i, j] = f(X[i, j], Y[i, j])`. A mesh can be created using the `np.meshgrid` function in Python. The `meshgrid` function has the inputs $x$ and $y$, which are lists containing the independent dataset. The output variables `X` and `Y` were described earlier.  

**Try it!**: Create a mesh for `x = [1, 2, 3, 4]` and `y = [3, 4, 5]` using the `meshgrid` function.  

```python
x = [1, 2, 3, 4]
y = [3, 4, 5]

X, Y = np.meshgrid(x, y)

print(X)
print(Y)
```

In addition, we can plot 3D surfaces in Python: the function to plot 3D surfaces is `plot_surface(X, Y, Z)`, where `X` and `Y` are the output arrays from `meshgrid`, and `Z = f(X, Y)` or `Z[i, j] = f(X[i, j], Y[i, j])`.

**Try it!**: make a plot of the surface $f(x, y) = sin(x)\cdot cos(x)$ for $-5≤x≤5, -5≤y≤5$ using the `plot_surface` function. take care to use a sufficiently fine discretization in `x` and `y` so that the plot looks smooth.

```python
fig = plt.figure(figsize=(12, 10))
ax = plt.axes(projection = "3d")

x = np.arange(-5, 5.1, 0.2)
y = np.arange(-5, 5.1, 0.2)

X, Y = np.meshgrid(x, y)
Z = np.sin(X)*np.cos(Y)

surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)

# Set axes label
ax.set_xlabel("x", labelpad=20)
ax.set_ylabel("y", labelpad=20)
ax.set_zlabel("z", labelpad=20)

fig.colorbar(surf, shrink = 0.5, aspect = 8)

plt.show()
```

We can have subplots of different 3D plots as well, To do that, use the `add_subplot` function from the figure object we created to generate the subplots for 3D cases.

**Try it!**: Make a $1 × 2$ subplot to plot the above `X, Y, Z` data in a wireframe plot and a surface plot.  

```python
fig = plt.figure(figsize = (12, 6))

ax = fig.add_subplot(1, 2, 1, projection = "3d")
ax.plot_wireframe(X, Y, Z)
ax.set_title("Wireframe plot")

ax = fig.add_subplot(1, 2, 2, projection = "3d")
ax.plot_surface(X, Y, Z)
ax.set_title("Surface plot")

plt.tight_layout()

plt.show()
```

###**12.4 Animations and Movies**

We can create animations in Python by calling a plot function inside of a loop (usually a for-loop). The main tools for making animations in Python is the `matplolib.animation.Animation` base class, which provides a framework around which the animation functionality is built. See the example below.

**Try it!**: Create an animation of a red circle following a blue sine wave.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as manimation

n = 1000
x = np.linspace(0, 6*np.pi, n)
y = np.sin(x)

# Define the meta data for the movie
FFMpegWriter = manimation.writers["ffmpeg"]
metadata = dict(title = "Movie Test", artist = "Matplotlib",
                comment = "a red circle following a blue wave")
writer = FFMpegWriter(fps = 15, metadata= metadata)

# Initialize the movie
fig = plt.figure()

# plot the sine wave line
sine_line, = plt.plot(x, y, "b")
red_circle, = plt.plot([], [], "r", markersize = 10) # Unpack the list here
plt.xlabel("x")
plt.ylabel("sin(x)")

# update the frames for the movie

with writer.saving(fig, "writer_test.mp4", 100):

  for i in range(n):
      x0 = x[i]
      y0 = y[i]
      red_circle.set_data(x0, y0)
      writer.grab_frame()
```

4. Write a function `my_polygon(n)` that plots a regular polygon with `n` sides and radius 1. Recall that the radius of a regular polygon is the distance from its centroid to the vertices. Use `plt.axis("equal")` to make the polygon look regular. Remember to give the axes a label and a title. Use `plt.title` to title the plot according to the number of sides. **Hint:** This problem is significantly easier in terms of polar coordinates. Recall that a complete revolution around the unit circle is $2\pi$ radians. Note that the first and last point on the plygon should be the point associated with the polar coordinate angles, $0$ and $2\pi$, respectively.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
#Define the function
# Define the radius
r = 1
def my_polygon(n):
  """
  Function that plots a regular polygon of n sides
  """

  # Define the angle depending on the number of sides (n)
  angle = np.arange(0, 2*math.pi + 2*math.pi/n, 2*math.pi/n)
  x_coord = r*np.cos(angle)
  y_coord = r*np.sin(angle)
  plt.figure()
  plt.axis("equal")
  plt.plot(x_coord, y_coord)
  plt.title(f"My Regular Polygone with {n} sides")
  plt.xlabel("x", labelpad=20)
  plt.ylabel("y", labelpad=20)
  plt.show();
```

```python
import numpy as np
import matplotlib.pyplot as plt

def my_fun_plotter(f, x):
  """
  Function that plots a lambda function (f) evaluated at x
  """
  plt.figure()
  y = f(x) # Evaluate the function f with input x
  plt.plot(x, y) # Plot x against the evaluated y
  plt.xlabel("x", labelpad=20)
  plt.ylabel("y", labelpad=20)
  plt.show();
```

6. Write a function with `my_poly_plotter(n, x)`  that plots the polynomials $p_{k}(x)=x^{k}$ for $k = 1, ..., n$. Make sure your plot has axis labels and a title.

```python
def my_poly_plotter(n, x):
  """
  Function that plots polynomials up to degree n
  """
  for i in np.arange(1, n + 1, 1):
    y = x**i # Create the polynomial function
    plt.plot(x, y) # Plot x against the evaluated y
    plt.xlabel("x", labelpad=20)
    plt.ylabel("y", labelpad=20)
    plt.title(f"Polynomials up to degree {n}")
  plt.show();
```

Write a function `my_sirpinski(n)` that generates the points $p_{i}$ for $i = 1, ..., n$. The function should make a plot of the pints using blue dots (i.e., "`b.`" as the third argument to the `plt.plot` function).

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

# Triangle vertices
p1 = [0, 0]
p2 = [0.5, 2**0.5/2]
p3 = [1, 0]

def my_sierpinski(n):
  """
  Generate and plot points following the Sierpiński triangle algorithm
  """

  # Start at point p1
  current_point = np.array(p1)

  # Empty vector for storing updated coordinates
  x_vals = []
  y_vals = []

  # Vertices to choose from:
  vertices = [np.array(p1), np.array(p2), np.array(p3)]

  # Generate points
  for _ in range(n):
    # Choose a random vertex
    chosen_vertex = random.choice(vertices)

    # Compute the midpoint between current point and chosen vertex
    current_point = (current_point + chosen_vertex)/2

    # Store point
    x_vals.append(current_point[0])
    y_vals.append(current_point[1])

  # Plot all generated points
  plt.figure(figsize = (6, 6))
  plt.scatter(x_vals, y_vals, s = 1)
  plt.title(f"Sierpinski Triangle (n = {n})")
  plt.axis("equal")
  #plt.axis("off")
  plt.show()
```

```python
# Test case 1:
my_sierpinski(100)
```

```python
# Test case 2:
my_sierpinski(10000)
```

8. Assume you are generating a set of points $(x_{i}, y_{i})$ where $x_{1}=0$ and $y_{1}=0$. The points $(x_{i}, y_{i})$ for $i = 2, ..., n$ are generated according to the following probabilistic relationship:

With 1% probability:

$x_{i} = 0$,

$y_{i}=0.16y_{i-1}$;

With 7% probability:

$x_{i}=0.2x_{i-1}-0.26y_{i-1}$,

$y_{i}=0.23x_{i-1}+0.22y_{i-1} + 1.6$;

With 7% probability:

$x_{i}=-0.15x_{i-1}+0.28y_{i-1}$,

$y_{i}=0.26x_{i-1}+0.24y_{i-1} + 0.44$;

With 85% probability:

$x_{i}=0.85x_{i-1}+0.04y_{i-1}$,

$y_{i}=-0.04x_{i-1}+0.85y_{i-1} + 1.6$;

Write a function `my_fern(n)` that generates the points $(x_{i}, y_{i})$ for $i = 1, ..., n$ and plots them using blue dots. Also use `plt.axis("equal")` and `plt.axis("off")` to make the plot more attractive.  





```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

def my_fern(n):
  """
  Stochastic dots generator (fractal)
  """
  # Initial coordinates
  x, y = 0, 0

  # Empty vectors to store the generated coordinates
  x_val = []
  y_val = []

  # Transformations and their probabilities
  transformations = [
      lambda x, y: (0, 0.16 * y),
      lambda x, y: (0.2 * x - 0.26 * y, 0.23 * x + 0.22 * y + 1.6),
      lambda x, y: (-0.15 * x + 0.28 * y, 0.26 * x + 0.24 * y + 0.44),
      lambda x, y: (0.85 * x + 0.04 * y, -0.04 * x + 0.85 * y + 1.6)
  ]
  probabilities = [0.01, 0.07, 0.07, 0.85]


  # Generate points
  for _ in range(n):
    # Choose a transformation based on probabilities
    chosen_transform = random.choices(transformations, weights=probabilities, k=1)[0]

    # Apply the transformation
    x, y = chosen_transform(x, y)

    # Store point
    x_val.append(x)
    y_val.append(y)

  # Plot all generated points
  plt.figure(figsize = (6, 6))
  plt.scatter(x_val, y_val, s = 1, c = "blue")
  plt.title(f"My fern with {n} iterations")
  plt.axis("equal")
  plt.axis("off")
  plt.show()
```

```python
# Test case 1:
my_fern(100)
```

```python
# Test case 2:
my_fern(10000)
```
