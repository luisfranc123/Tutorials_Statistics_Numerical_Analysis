####**21.1 NUMERICAL INTEGRATION PROBLEM STATEMENT**

Given a function $f(x)$, approximate the integral of $f(x)$ over the total **interval**, $[a, b]$. To accomplish this goal, weassume that the interval has been discretized into a numeral grid, $x$, consisting of $n + 1$ points with spacing, $h = \frac{b-a}{n}$. Here, we denote each point in $x$ by $x_i$, where $x_0=a$ and $x_n = b$. Note that there are $n + 1$ grid points because thecount starts at $x_0$. Assume that the function, $f(x)$, can be computed for any of the grid points, or that we have been given the function implicitly as $f(x_i)$. The interval $[x_i, x_{i+1}]$ is referred to as a **subinterval**.

####**21.2 RIEMANN INTEGRAL**

The simplest method for approximating integrals is by summing the area of rectangles that are defined for each subinterval. The width if the rectangle is $x_{i+1} - x_i = h$, and the height is defined by a function value $f(x)$ for some $x$in the subinterval. An obvious choice for the height is the function value at the left endpoint, $x_i$, or the right endpoint, $x_{i+1}$, because these values can be used even is the function itself is not known. This method gives the **Riemann integral** approximation, which is

$$\int^{b}_{a} {f(x)dx} \approx \sum_{i = 0}^{n-1}{hf(x_i)},$$

or

$$\int^{b}_{a} {f(x)dx} \approx \sum_{i = 1}^{n}{hf(x_i)},$$

depending on whether the left or right endpoing is chosen.

As with numerical differenciation, we want to characterize how accuracy improves as $h$ decreases. To determine this characterizing, we first rewrite the integral of $f(x)$ over an arbitrary subinterval in terms of the Taylor series. The Taylor series of $f(x)$ around $a = x_i$ is

$$f(x) = f(x_i) + f^{'}(x_i)(x-x_i) + ...$$

Thus

$$\int_{x_i}^{x_{i+1}}{f(x)dx} = \int_{x_i}^{x_{i+1}}{(f(x_i) + f^{'}(x_i)(x - x_i) + ...)dx}$$

by substitution of the Taylor series for the finction. Since the integral distributes, we can rearrange the right-hand side into the following form:

$$\int_{x_i}^{x_{i+1}}{f(x_i)dx} + \int_{x_i}^{x_{i+1}}{f^{'}(x_i)(x - x_i)dx + ...}$$

Computing each integral separately results in the approximation

$$\int_{x_i}^{x_{i+1}}{f(x)dx} = hf(x_i) + \frac{h^2}{2}f^{'}(x_i) + O(h^3),$$

which is

$$\int_{x_i}^{x_{i+1}}{f(x)dx} = ff(x_i) + O(h^2).$$

Since the $hf(x_i)$ term is our Riemann integral approximation for a single subinterval, the Riemann integral approximation over a single interval is $O(h^2)$.

If we sum the $O(h^2)$ error over the entire Riemann sum, we obtain $n\space{O(h^2)}$.The relationship between $n$ and $h$ is

$$h = \frac{b-a}{n},$$

and so our total error becomes $\frac{b-a}{n}O(h^2) = O(h)$ over the whole interval. Thus the overall accuracy is $O(h)$.

The **Midpoint Rule** takes the rectangle height of the rectangle at each subinterval to be the function value at the midpoint between $x_i$ and $x_{i+1}$, which for compactness we denote by $y_i = \frac{x_{i+1}+x_i}{2}$. The midpoint rule says

$$\int^{b}_{a} {f(x)dx} \approx \sum_{i = 0}^{n-1}{hf(y_i)}.$$

Similarly to the Riemann integral, we take the Taylor series of $f(x)$ around $y_i$, which is

$$f(x) = f(y_i) + f^{'}(y_i)(x-y_i) + \frac{f^{''}(y_i)(x-y_i)^2}{2!} + ...$$

Then the integral over a subinterval is

$$\int^{x_{i+1}}_{x_i} {f(x)dx} = \int^{x_{i+1}}_{x_i} {(f(y_i) + f^{'}(y_i)(x-y_i) + \frac{f^{''}(y_i)(x-y_i)^2}{2!} + ...)dx},$$

which distributes to

$$\int^{x_{i+1}}_{x_i} {f(x)dx} = \int^{x_{i+1}}_{x_i}{f(y_i)dx + \int^{x_{i+1}}_{x_i}f^{'}(y_i)(x-y_i)dx + \int^{x_{i+1}}_{x_i}\frac{f^{''}(y_i)(x-y_i)^2}{2!}dx + ...}$$

Recognizing that $x_i$ and $x_{i+1}$ are symmetric around $y_i$, we get $\int^{x_{i+1}}_{x_i}f^{'}(y_i)(x-y_i)dx = 0$. This is true for the interval of $(x-y_i)^p$ for any odd $p$. For the integral of $(x-y_i)^p$ and with $p$ even, $\int^{x_{i+1}}_{x_i}(x-y_i)^pdx = \int^{{h/2}}_{-h/2}x^pdx$, which will result in some multiple of $h^{p+1}$ with no lower order powers of $h$.

Utilizing these facts the expression for the integral of $f(x)$ to

$$\int^{x_{i+1}}_{x_i}{f(x)dx} = hf(y_i) + O(h^3).$$

Since $hf(y_i)$ is the approximation of the integral over the subinterval, the midpoint rule is $O(h^3)$ for one subinterval; using tsimilar arguments as those used for the Riemann integral, we get $O(h^2)$ over the whole intervl. Since the idpoint rule rquires the same number of calculations as the Riemann integral, we essentially get an extra order of accuracy for free; however, if $f(x_i)$ is given in the form of data points, then we will not be able to compute $f(y_i)$ for this integration scheme.

**Try it!** Use the left and right Riemann integral, as well as midpoint rule, to approximate $\int_{0}^{\pi}sin(x)dx$ with 11 evenly spaced grid points over the whole interval. Compare this value to the exact value of 2.




```python
import numpy as np

# Initialize parameters
a = 0
b = np.pi
n = 11
h = (b-a)/(n-1)
x = np.linspace(a, b, n)
# Midpoint computation
y_i = (x[:n-1] + x[1:])/2
f = np.sin(x)

 # Riemann approximation (left)
I_riemannL = h * sum(f[:n-1])
err_riemannL = 2 - I_riemannL

 # Riemann approximation (right)
I_riemannR = h * sum(f[1:])
err_riemannR = 2 - I_riemannR

 # Approximation (midpoint)
I_mid = h * sum(np.sin(y_i))
err_mid = 2 - I_mid

print(I_riemannL)
print(err_riemannL)
print("-------------------------------")
print(I_riemannR)
print(err_riemannR)
print("-------------------------------")
print(I_mid)
print(err_mid)
```

####**21.3 TRAPEZOID RULE**

The **Trapezoid Rule** fits a trepzoid into each subinterval and sums the areas of the trepzoids to approximate the total integral.This approximation for the integral to an arbitrary function is shown in the figure below. For each subinterval, the trapezoid rule computes the area of a trapezoid with corners at $(x_i, 0), (x_{i+1}, 0)$, and $(x_{i+1}, f(x_{i+1}))$, which is $h\frac{f(x_i)+f(x_{i+1})}{2}$. This, the trapezoid rule approximates integrals according to the expression

$$\int_{a}^b {f(x)dx\approx \sum_{i=0}^{n-1}{h\frac{f(x_i)+f(x_{i+1})}{2}}}.$$

<img src="https://pythonnumericalmethods.studentorg.berkeley.edu/_images/21.03.1-Trapezoid_integral.png" width="400" height="250">

**TRY IT!** You may notice that the trapezoid rule "double-counts" most of the terms in the series. To illustrate this fact, consider the exmansion of the trapezoid rule:

$$\sum_{i=0}^{n-1}{h\frac{f(x_i)+f(x_{i+1})}{2}} = \frac{h}{2}[(f(x_0)+f(x_1)) + (f(x_1) + f(x_2) + f(x_3)) + ... + (f(x_{n-1}) + f(x_n))].$$

Computationally, this is many extra additions and calls to $f(x)$ than are really necessary. We can be made more computationally efficient using the following expression:

$$\int_{a}^{b}{f(x)dx} \approx \frac{h}{2}(f(x_0) + 2\sum_{i=1}^{n-1}{f(x_i) + f(x_n)}).$$






To determine the accuracy of the trapezoid rule approximation, we first take Taylor series expansion of $f(x)$ around $y_i=\frac{x_{i+1}+x_i}{2}$, which is the midpoint between $x_i$ nd $x_{i+1}$. This Taylor series expansion is

$$f(x) = f(y_i) + f^{'}(y_1)(x - x_i) + \frac{f^{''}(y_1)(x-y_i)^2}{2!} + ...$$

Computing the Taylor series at $x_{i+1}$ and noting that $x_i-y_i = -\frac{h}{2}$ and $x_{i+1}-y_i = \frac{h}{2}$ results in the following expressions:

$$f(x_i) = f(y_i) - \frac{hf^{'}(y_i)}{2} + \frac{h^2f^{''}(y_i)}{8} - ...$$

and

$$f(x_{i+1}) = f(y_i) + \frac{hf^{'}(y_i)}{2} + \frac{h^2f^{''}(y_i)}{8} + ...$$

Taking the average of these two expressions results in the new expression,

$$\frac{f(x_{i+1}) + f(x_i)}{2} = f(y_i) + O(h^2).$$

Solving this expression for $f(y_i)$ yields

$$f(y_i) = \frac{f(x_{i+1}) + f(x_i)}{2} + O(h^2).$$

Now returning to the Taylor expansion for $f(x)$, the integral of $f(x)$ over a subinterval is

$$\int_{x_i}^{x_{i+1}}{f(x)dx} = \int_{x_i}^{x_{i+1}}{(f(x_i)+f^{'}(y_i)(x-y_i) + \frac{f^{''}(y_i)(x-y_i)^2}{2!} + ...)dx}.$$

Distributing the itegral results in the expression

$$\int_{x_i}^{x_{i+1}}{f(x)dx} = \int_{x_i}^{x_{i+1}}{f(y_i)dx} + \int_{x_i}^{x_{i+1}}{f^{'}(y_i)(x-y_i)dx + \int_{x_i}^{x_{i+1}}{\frac{f^{''}(y_i)(x-y_i)^2}{2!}}dx} + ...$$

Since $x_i$ and $x_{i+1}$ are symmetric around $y_i$, the integrals of the odd powers of $(x-y_i)^p$ disappear, and the even powers resolve to a multiple $h^{p+1}$:

$$\int_{x_i}^{x_{i+1}}{f(x)dx} = hf(y_i) + O(h^3).$$

If we substitute $f(y_i)$ with the expression derived explictly i terms of $f(x_i)$ and $f(x_{i+1})$, we obtain

$$\int_{x_i}^{x_{i+1}}{f(x)dx} = h(\frac{f(x_{i+1} + f(x_i))}{2} + O(h^2)) + O(h^3),$$


which is equivalent to

$$h(\frac{f(x_{i+1} + f(x_i))}{2}) + hO(h^2) + O(h^3)$$

and

$$\int_{x_i}^{x_{i+1}}{f(x)dx} = h(\frac{f(x_{i+1} + f(x_i))}{2}) + O(h^3).$$








Since $\frac{h}{2}(f(x_i+1) + f(x_i))$ is the trapezoid rule approximation for the integral over the subinterval, it is $O(h^3)$ for a single subinterval and $O(h^2)$ over the whle interval.

**TRY IT!** Use the trapezoid rule to approxmate $\int_{0}^{\pi}{sin(x)dx}$ with 11 evenly spaced points over the whole interval. Compare this value to the exact value of 2.

```python
import numpy as np

# Initialize parameters

a = 0
b = np.pi
n = 11
h = (b-a)/(n-1)
x = np.linspace(a, b, n)
f = lambda x: np.sin(x)

I_trap = (h/2)*(f(x[a]) + 2*sum(f(x[1:n-1])) + f(x[n-1]))
err_trap = 2 - I_trap

print(I_trap)
print("------")
print(err_trap)
```

####**21.4 SIMPSON'S RULE**

Consider *two* consecutive subintervals, $[x_{i-1}, xi]$ and $[x_i, x_{i+1}]$. **Simpson's Rule** approximates the area under $f(x)$ over these two subintervals by fitting a quadratic polynomial through the points $(x_{i-1}, f(x_{i-1}))$, $(x_i, f(x_i))$, and $(x_{i+1}, f(x_{i+1}))$, which is a unique polynomial, and then integrates the quadratic exactly. Figure below shows this integral approximation for an arbitrary function.

<img src="https://pythonnumericalmethods.studentorg.berkeley.edu/_images/21.04.1-Simpson_integral.png" width="400" height="250">

First, we construct the quadratic polynomial approximation of the function over the two subintervals. The easiest way to do this is to use Lagrange polynomials. By applying the formula for constructing Lagrange polynomials, we obtain

$$p_i(x)=f(x_{i-1})\frac{(x-x_i)(x-x_{i+1})}{(x_{i-1}-x_i)(x_{i-1}-x_{i+1})} + f(x_{i})\frac{(x-x_{i-1})(x-x_{i+1})}{(x_i-x_{i-1})(x_i-x_{i+1})} + f(x_{i+1})\frac{(x-x_{i-1})(x-x_i)}{(x_{i+1}-x_i)(x_{i+1}-x_{i-1})}$$

and with substitutions for $h$ results in

$$p_i(x)=\frac{f(x_{i-1})}{2h^2}(x-x_i)(x-x_{i+1})-\frac{f(x_i)}{h^2}(x-x_{i-1})(x-x_{i+1}) + \frac{f(x_{i+1})}{2h^2}(x-x_{i-1})(x-x_i).$$

We can confirm tht the polynomial curve intersects the desired points. With some algebra and manipulation, the integral of $p_i(x)$ over the two subintervals is

$$\int_{x_{i-1}}^{x_{x_{i+1}}}{p_i(x)dx} = \frac{h}{3}(f(x_{i-1}) + 4f(x_i) + f(x_{i+1})).$$

To approximate the integral over $(a, b)$, we must sum the integrals of $P_i(x)$ over all *pairs* of subintervals since $P_i(x)$ spans two subintervals. Substituting $\frac{h}{3}(f(x_{i-1}) + 4f(x_i) + f(x_{i+1}))$ for the integral of $P_i(x)$ and regrouping the terms for efficiency leads to the formula

$$\int_{a}^{b}{f(x)dx} \approx \frac{h}{3}[f(x_0) + 4\sum_{i=1, i\space{odd}}^{n-1}{f(x_i)}+2\sum_{i=2, i\space even}^{n-2}{f(x_i)} + f(x_n)].$$

The regrouping is illustrated in the figure below:

<img src="https://ars.els-cdn.com/content/image/3-s2.0-B9780128195499000312-f21-28-9780128195499.jpg" width="400" height="250">

**WARNING!** Note that to use Simpson's rule, we **must** have an even number of intervals and, therefore, an odd number of grid points.

####**21.5 COMPUTING INTEGRALS IN PYTHON**

The `scipy.integrate` function subpackage has several functions for computing integrals. The `trapezoid` takes as iput arguments an array of function values $f$ computed on a numerical grid $x$.

**TRY IT!** Use the `trapezoid` function to approximate $\int_{0}^{\pi}{sin(x)dx}$ for 11 evenly spaced points over the whole interval. Compare this value to the one computed in the earlier example using the trapezoid rule.

```python
import numpy as np
from scipy.integrate import trapezoid

# Initialize the parameters
a = 0
b = np.pi
n = 11
h = (b-a)/(n-1)
x = np.linspace(a, b, n)
f = lambda x: np.sin(x)

I_trapz = trapezoid(f(x), x)
I_trap = (h/2)*(f(x[0]) +2*sum(f(x[1:n-1])) + f(x[n-1]))

print(I_trapz)
print(I_trap)
print(f"error = {(abs(I_trapz - I_trap)/I_trapz)*100}")
```

Sometimes we need to know the approximated cumulative integral. That is, $F(X)=\int_{0}^{\pi}{f(x)dx}$. For this purpose, it is useful to use the `cumulative_trapezoid` function, which takes the same input arguments as `trapezoid`.

**TRY IT!** Use the `cumulative_trapezoid` function to approximate the cumulative integral of $f(x) = sin(x)$ from $0$ to $\pi$, with a discretization step of 0.01. The exact solution of this integral is $F(x) = sin(x)$. Plot the results.  

```python
from scipy.integrate import cumulative_trapezoid
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-notebook')

x = np.arange(0, np.pi, 0.01)
F_exact = -np.cos(x)
F_approx = cumulative_trapezoid(np.sin(x), x)

plt.figure(figsize = (10, 6))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title("$F(x) = \int_0^{x}{sin(y)dy}$");
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["Exact with Offset", "Approx"])
plt.show();
```

**1.** Write a function `my_int_calc(f, f0, a, b, N, option)` where `f` is a function object, `a` and `b` are scalars such that `a < b`, `N` is a positive integer, and `option` is the string `"rect"`, `"trap"`, or `"simp"`. Let `x` be an array starting at `a`, ending at `b`, and containing `N` evenly spaced elemtns. The output argument, `I`, should be an approximation to the integral of `f(x)`, with initial condition `f0` computed according to the input argument, `option`.



```python
import numpy as np
def my_int_calc(f, f0, a, b, N, option):

  # Initialize parameters
  x = np.linspace(a, b, N)
  h = (b-a)/(N-1)
  f0 = f(x[0])

  # Riemann integral approximation
  if option == "rect":
    I_riemannL = h*sum(f(x[:N-1]))
    I_riemannR = h*sum(f(x[1:]))
    I = (I_riemannL + I_riemannL)/2

  # Trapezoid integral approximation
  elif option == "trap":
    I = (h/2)*(f0 + 2*sum(f(x[1:N-1])) + f(x[N-1]))

  # Simpson's rule integral approximation
  elif option == "simp":
    I = (h/3)*(f0 + 4*sum(f(x[1:N-1:2])) + 2*sum(f(x[:N-2:2])) + f(x[N-1]))

  # Raise error if option not within the options
  else:
    raise ValueError("option must be 'rect', 'trap', or 'simp'")

  return I
```

```python
from scipy.integrate import quad

f = lambda x: np.sin(np.exp(x))
a = 0
b = np.pi
N = 21
print(f"rect_approximation = {my_int_calc(f, f(a), a, b, N, "rect")}")
print(f"trap_approximation = {my_int_calc(f, f(a), a, b, N, "trap")}")
print(f"simp_approximation = {my_int_calc(f, f(a), a, b, N, "simp")}")
quad_int, err_quad = quad(f, a, b)
print(f"quad approximation = {quad_int}")
```

**2.** An earlier chapter demosntrated that some functions can be expressed as an infinite sum of polynomials (i.e., the Taylor series). Other functions, particularly periodic functions, can be written as an infinite sum of sine and cosine waves. For these functions,

$$f(x)=\frac{A_0}{2} + \sum_{n = 1}^{\infty}{A_ncos(nx) + B_nsin(nx)}$$

It can be shown that the values of $A_n$ and $B_n$ can be computed using the following formulas:

$$A_n = \frac{1}{\pi}\int_{-\pi}^{\pi}{f(x)\cdot cos(nx)\space{dx}},$$

$$B_n = \frac{1}{\pi}\int_{-\pi}^{\pi}{f(x)\cdot sin(nx)\space{dx}}.$$

Just like the Taylor series, functions can be approximated by truncating the Furies series at some $n = N$. Fourier series can be used to approximate some particularly nasty functions, such as the step function, and they form the basis of many engineering applications, such as signal processing. Write a function `my_fourier_coef(f, n)` with output `[An, Bn]`, where `f` is a function object that is $2\pi$-periodic. The function `my_fourier_coef` should compute the $n$th Fourier coefficients, `An` and `Bn`, in the Fourier series for `f` defined by the two formulas given earlier. Use the *quad* function to perform the integration.



```python
from scipy.integrate import quad
import numpy as np

def my_fourier_coef(f, n):
  """
  Function to calculate the Fourier
  coefficients truncated to n
  """
  f_An = lambda x: np.cos(n*x)
  f_Bn = lambda x: np.sin(n*x)

  # Define new lambda functions for the product of f and f_An or f_Bn
  integrand_An = lambda x: f(x) * f_An(x)
  integrand_Bn = lambda x: f(x) * f_Bn(x)

  I_An, est_err_quad_An = quad(integrand_An, -np.pi, np.pi)
  I_Bn, est_err_quad_Bn = quad(integrand_Bn, -np.pi, np.pi)
  An = (1/np.pi)*I_An
  Bn = (1/np.pi)*I_Bn

  return [An, Bn]
```

```python
# Test case 1
def plot_results(f, N):
  x = np.linspace(-np.pi, np.pi, 10000)
  [A0, B0] = my_fourier_coef(f, 0)
  y = A0*np.ones(len(x))/2

  for n in range(1, N):
    [An, Bn] = my_fourier_coef(f, n)
    y += An*np.cos(n*x)+Bn*np.sin(n*x)
  plt.figure(figsize = (10,6))
  plt.plot(x, f(x), label = "analytic")
  plt.plot(x, y, label = "approximate")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.grid()
  plt.legend()
  plt.title(f"{N}th Order Fourier Approximation")
  plt.show()

f = lambda x: np.sin(np.exp(x))
N = 2
plot_results(f, N)
```
