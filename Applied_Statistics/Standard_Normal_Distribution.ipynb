##**1. Functions to compute *Z-score* and calculate probabilities**
---

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to calculate the Z-score
def z_score(value, mean, std_dev):
    return (value - mean) / std_dev

# Function to calculate probabilities based on Z-scores and normal distribution
def normal_prob(lower_bound, upper_bound, mean, std_dev):
    lower_z = z_score(lower_bound, mean, std_dev)
    upper_z = z_score(upper_bound, mean, std_dev)
    lower_prob = stats.norm.cdf(lower_z)
    upper_prob = stats.norm.cdf(upper_z)
    return upper_prob - lower_prob
```

###*Example 2: Cholesterol levels*
---
The cholesterol levels of adults in a certain population are normally distributed with a mean of 200 mg/dL and a standard deviation of 30 mg/dL. A health study is interested in knowing the probability that an adult will have a cholesterol level either below 150 mg/dL or above 250 mg/dL.

```python
# Example 2: Cholesterol levels
mean_cholesterol = 200   # mg/dL
std_cholesterol = 30     # mg/dL
lower_cholesterol = 150  # mg/dL
upper_cholesterol = 250  # mg/dL

prob_cholesterol_lower = stats.norm.cdf(z_score(lower_cholesterol, mean_cholesterol, std_cholesterol))
prob_cholesterol_upper = 1 - stats.norm.cdf(z_score(upper_cholesterol, mean_cholesterol, std_cholesterol))
prob_cholesterol = prob_cholesterol_lower + prob_cholesterol_upper

print(f"Probability of cholesterol level below {lower_cholesterol} or above {upper_cholesterol} mg/dL: {prob_cholesterol:.4f}")
```
